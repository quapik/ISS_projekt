```python
%matplotlib notebook
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import IPython
data,fs=sf.read('off.wav')
data-=np.mean(data)
data1,fs=sf.read('on.wav')
data1-=np.mean(data1)

```

```python
counter=0
x=[]
for i in range(0,99):
    y=[]
    for j in range(0,320):
        y.append(data[j+160*counter])
      
    x.append(y)
    counter=counter+1
counter=0
x2=[]
for i in range(0,99):
    y=[]
    for j in range(0,320):
        y.append(data1[j+160*counter])
        
    x2.append(y)
    counter=counter+1
plt.figure(figsize=(10,3))
plt.title("Porovnn dvou rznch rmc bez rouky a s roukou")
plt.plot(x[0],label='MaskOFF')
plt.plot(x2[0],label='MaskON')
plt.xlabel('Poet vzork(1 rmec)')
plt.legend()
```
